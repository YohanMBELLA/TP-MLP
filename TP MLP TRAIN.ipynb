```python
import pandas as pd
import numpy as np


iden=pd.read_csv("train.csv")

iden.head()
```

```python

train=iden.replace(['male','female',np.nan],[1,0,-5])
```

```python
train
```

```python
v=train.values
print(v)
```

```python
tab=[]
count=0

for i in range (6):
    for j in range (891):
        if v[j,i]==-5:
            count=j
            tab.append(count)
            
            
print(tab)            
```

```python
M=np.delete(v,(tab),axis=0)
M.shape
```

```python
Xi=M[:,[2,4,5,6]]

y=M[:,1]

y=y.astype('int')
```

```python
import matplotlib.pyplot as plt

import time
from sklearn.model_selection import  GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(Xi)
X=scaler.transform(Xi)


# Split data into train partition and test partition
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)

start_time=time.time()


mlp = MLPClassifier( hidden_layer_sizes=(70),max_iter=60,alpha=1e-3,learning_rate='adaptive')
mlp.fit(X_train,y_train)

end_time=time.time()

print((end_time)-(start_time))

print(mlp.score(X_test,y_test))
```
